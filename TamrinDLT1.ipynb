```python
# Ex.5
torch.cuda.manual_seed(1234)
a = torch.randn(7,7, requires_grad = True, device = device)
b = torch.rand(1,7, requires_grad = False, device = device)
c = a@(b.T)
c
```

```python
# Ex.6
torch.manual_seed(1234)
a = torch.randn(2,3, device = 'cpu')
b = torch.randn(2,3, device = 'cpu')
a.to('cuda')
b.to('cuda')
```

```python
print(a)
print(b)
```

```python
# Ex.7
b2 = b.clone().detach().view(3,2)
b2
```

```python
c = a@b2
c
```

```python
# Ex.8
min = torch.min(c)
min
```

```python
max = torch.max(c)
max
```

```python
#Ex.9
imax = torch.argmax(c)
imin  = torch.argmin(c)
print(imax, imin)
```

```python
max_index_ndim = torch.unravel_index(torch.argmin(c), c.shape)
print(max_index_ndim)
```
